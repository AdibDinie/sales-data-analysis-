```python
# Total sales by branch
branch_sales = df.groupby('Branch')['Total'].sum().sort_values(ascending=False)

# Plotting
branch_sales.plot(kind='bar', title='Total Sales by Branch', ylabel='Total Sales', xlabel='Branch', color='skyblue')
plt.xticks(rotation=0)
plt.show()
```

```python
# Total sales by product line
product_sales = df.groupby('Product line')['Total'].sum().sort_values()

# Plotting
product_sales.plot(kind='barh', figsize=(8,5), title='Total Sales by Product Line', color='salmon')
plt.xlabel('Total Sales')
plt.show()
```

```python
gender_sales = df.groupby('Gender')['Total'].sum()

# Plot
gender_sales.plot(kind='bar', title='Total Sales by Gender', ylabel='Total Sales', xlabel='Gender', color=['purple', 'orange'])
plt.xticks(rotation=0)
plt.show()
```

```python
# Count payment methods
payment_counts = df['Payment'].value_counts()

# Plotting
payment_counts.plot(kind='pie', autopct='%1.1f%%', title='Payment Method Distribution', figsize=(6,6))
plt.ylabel('')
plt.show()
```

```python
# Convert Time column to datetime and extract Hour
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M')
df['Hour'] = df['Time'].dt.hour
```

```python
# Total sales by hour
hourly_sales = df.groupby('Hour')['Total'].sum()

# Plotting
plt.figure(figsize=(10,5))
hourly_sales.plot(kind='bar', color='teal')
plt.title('Total Sales by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Day'] = df['Date'].dt.day_name()

```

```python
# Sales by day of the week
day_sales = df.groupby('Day')['Total'].sum().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

# Plotting
plt.figure(figsize=(10,5))
day_sales.plot(kind='bar', color='coral')
plt.title('Total Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Average spending per transaction by customer type
avg_spending = df.groupby('Customer type')['Total'].mean()

# Plot
avg_spending.plot(kind='bar', color='lightseagreen', title='Avg Spending by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Average Total')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Pivot table: Gender vs Customer Type
pivot = df.pivot_table(index='Gender', columns='Customer type', values='Total', aggfunc='mean')

# Plotting
pivot.plot(kind='bar', figsize=(8,5), title='Average Spending by Gender & Customer Type', colormap='viridis')
plt.ylabel('Average Total')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```
